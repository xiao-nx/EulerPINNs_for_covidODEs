```python
# -*-coding: utf-8 -*-
import os
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker
from matplotlib.pyplot import MultipleLocator

%matplotlib inline
```

```python
data_path = './data/'
fname = os.path.join(data_path, 'minnesota3.csv')
df = pd.read_csv(fname)
df.head()
```

```python
plt.figure(figsize=(20, 6))
confirmed = plt.bar(df['date'], df['cum_confirmed'], color='red', label='Cumulative confirmed cases')
recovered = plt.plot(df['date'], df['recovered'], color='green', label='Cumulative recovered cases')
death = plt.bar(df['date'], df['death'], color='black', label='Cumulative death cases')
#
plt.xlabel('Date',fontsize=14)
# 
plt.xticks(rotation=45)
#x
plt.ylabel('Cumulative Cases',fontsize=14)
plt.legend(loc='best')
plt.title('Cumulative confirmed cases in minnesota, the US', fontsize=24)
#
# plt.style.use({'figure.figsize':(10, 6)})
#
x_major_locator= MultipleLocator(10)#x1
y_major_locator= MultipleLocator(20000)#y5
ax= plt.gca()#ax
ax.xaxis.set_major_locator(x_major_locator)#x1
ax.yaxis.set_major_locator(y_major_locator)#y5

# plt.savefig('./figures/total_confirmed.png')
plt.show(block=False)
```

```python
df['state_confiremed'] = df['cum_confirmed'] - df['recovered'] - df['death']
```
